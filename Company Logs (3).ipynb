### 1.2.4 Terminology

### 2.1.3 Selection criteria
As already mentioned in point 2.1.2, all features in the dataset are relevant for achieving our data mining goals (developing a predictive model and analysing developper efficiency). However, for developping the model, we will not use the ID columns, as well as features that pertain to the future (are not known at the time of assigning a task) and therefore cannot be used when applying the model (SignOff, ActionTimeStamp, WorkedMinutes). Also, we are not going to use the column "EstimateInMinutes" directly in our model. Instead, we are going to use it only after we have the model ready, in order to compare the values in this column to the model's predictions in order to evaluate whether the model's predictions are more accurate than the ones made when assigning the tasks. 

Also, for preparing data for the model, we are going to replace some of the features with constructed features. Namely, we will extract the year and month from the variable CreateTime in order to take into account differences over time. we are also going to add the No of assignees as a constructed feature (it is the number of people the same task is assigned to). Also, we are planning to encode the names of assignees (AssignedTo), their departments (AssigneeDept) and positions (ActivityTranslation) into one-hot vectors, so that all assignees a task has been assigned to could be taken into account. (In case of ActivityTranslation, we will split it first into two features - country and position). 

As regards case ranges, we would need to leave out tasks which have not been signed off yet (given that those tasks have not yet been finished, their actual duration cannot be compared to estimated duration, which renders them unuseful for our modelling). (Out of the 264,051 rows of data, the SignOff includes a nan value in 24,711 cases) We would also need to leave out cases where SignOff date is "01/01/1970 00:00" (there are 117 such cases). ## 2.4 Verifying data quality

In order to assess data quality, it would be useful to get an overview of the number of missing values in each of the columns.

```python
log.info()
```

**Columns that contain null values are: 'SignOff', 'CreatorDept', 'AssigneeDept', 'WorkedMinutes' and 'EstimateInMinutes'.**

The missing values of the variable **"EstimateInMinutes"** do not pose a problem, because as already pointed out, we are not going to use this variable for building the model, but only for measuring the model's efficiency. We do not need to measure it based on the entire dataset. The rows not containing this value are still usable for building the model. 

The rows with the value missing in the **'SignOff'** column (24,711 values) correspond to tasks which have not been completed yet. We would need to leave those rows out. As mentioned above, we would also need to leave out cases where SignOff date is "01/01/1970 00:00" (there are 117 such cases).

**'CreatorDept'** includes 806 missing values and **'AssigneeDept'** includes 120,924 missing values, which is a considerable proportion of the dataset. In order to explore this issue, let's compose a table of persons:


```python
#TABLE OF PERSONS:
#Replacing "nan" values with "-", to unable concatenation of columns:
log['Creator']=log['Creator'].fillna('-')
log['CreatorDept']=log['CreatorDept'].fillna('-')
log['AssignedTo']=log['AssignedTo'].fillna('-')
log['ActivityTranslation']=log['ActivityTranslation'].fillna('-')
log['AssigneeDept']=log['AssigneeDept'].fillna('-')

#PERSONS WHO HAVE BEEN ASSIGNING TASKS TO OTHERS:
log['tmp2']=log['Creator']+ ', '+ log['CreatorDept']    
assigners_list=list(set(list(log['tmp2']))) 
assigners_names=[]
assigner_dept=[]
for el in assigners_list:
    if type(el)==str: #'nan' value has type 'float' - must be left out
        content=el.split(', ')
        assigners_names.append(content[0])
        assigner_dept.append(content[1])
assigners=pd.DataFrame()
assigners['name']=assigners_names
assigners['dept']=assigner_dept
    
#PERSONS TO WHOME TASKS HAVE BEEN ASSIGNED:
log['tmp']=log['AssignedTo']+ ', '+ log['ActivityTranslation'] + ', '+ log['AssigneeDept']    
persons_list=list(set(list(log['tmp']))) 
names=[]
country=[]
position=[]
dept=[]
for el in persons_list:
    if type(el)==str: #'nan' value has type 'float' - must be left out
        content=el.split(', ')
        names.append(content[0])
        c=content[1].split(' ')[0]
        country.append(c)
        position.append(re.sub(c+' ', '', content[1]))
        dept.append(content[2])
persons=pd.DataFrame()
persons['name']=names
persons['country']=country
persons['dept']=dept
persons['position']=position
persons['assignee']=True

is_assigner=[]
for i, name in enumerate(names):
    if name in assigners_names:
        is_assigner.append(True)
    else:
        is_assigner.append(False)
persons['assigner']=is_assigner

#Assigners who are not assignees:
a=set(assigners_names)
b=set(names)
c=a-b
for name in c:
    idx=assigners_names.index(name)
    row=[assigners_names[idx], '-', assigner_dept[idx], '-', False, True]
    #print(row)
    tmp=pd.DataFrame([row], columns=persons.columns.values)
    persons=persons.append(tmp, ignore_index=True)
```
